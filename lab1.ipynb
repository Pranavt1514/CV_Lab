```python
def apply_transform(img, trans):
    h, w = img.shape
    corners = np.array([
        [0, 0, 1],
        [w, 0, 1],
        [0, h, 1],
        [w, h, 1]
    ])
    new_corners = corners @ trans.T
    x1 = int(np.min(new_corners[:,0]))
    x2 = int(np.max(new_corners[:,0]))
    y1 = int(np.min(new_corners[:,1]))
    y2 = int(np.max(new_corners[:,1]))

    new_w = x2 - x1
    new_h = y2 - y1
    inve = np.linalg.inv(trans)
    out = np.zeros((new_h, new_w), dtype=np.uint8)

    for y in range(new_h):
        for x in range(new_w):
            curr = np.array([x + x1, y + y1, 1]) @ inve.T
            currx, curry = curr[0], curr[1]
            sx = int(round(currx))
            sy = int(round(curry))

            if 0 <= sx and sx< w and 0 <= sy and sy < h:
                out[y, x] = img[sy, sx]

    cv2.imshow("img out", out)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

	
```

```python

def scale_image(img, sx, sy):
    matrix = np.array([
        [sx, 0, 0],
        [0, sy, 0],
        [0, 0, 1]
    ], dtype=float)
    return apply_transform(img, matrix)

def shear_image(img, shx=0, shy=0):
    matrix = np.array([
        [1, shx, 0],
        [shy, 1, 0],
        [0, 0, 1]
    ], dtype=float)
    return apply_transform(img, matrix)


def rotate_image(img, angle_degrees):
    angle = math.radians(angle_degrees)
    cos_a = math.cos(angle)
    sin_a = math.sin(angle)

    matrix = np.array([
        [cos_a, -sin_a, 0],
        [sin_a, cos_a, 0],
        [0, 0, 1]
    ], dtype=float)
    return apply_transform(img, matrix)


def reflect_image(img, axis='x'):
    if axis == 'x': 
        matrix = np.array([
            [1, 0, 0],
            [0, -1, 0],
            [0, 0, 1]
        ], dtype=float)
    else:
        matrix = np.array([
            [-1, 0, 0],
            [0, 1, 0],
            [0, 0, 1]
        ], dtype=float)
    return apply_transform(img, matrix)


scale_image(img, 1.5, 1.5)
shear_image(img, shx=0.5)
rotate_image(img, 45)
reflect_image(img, axis='y')


	
```
